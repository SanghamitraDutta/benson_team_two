```python
def Add_StRe(x): 
    name = x
    if ('Broadway' in x):
        name = 'Broadway'
    if (x == '5th Ave') or (x=='Fifth Avenue'):
        name = '5th Avenue'
    if (x== 'Madison Ave') or (x== 'Madison Ave.'):
        name = 'Madison Avenue'
    if (x== 'Jay St.') or (x== 'Jay St') or (x=='Jay st.'):
        name = 'Jay Street'
    if ('Varick' in x):
        name = 'Varick Street'
    if (x== 'Water St'):
        name = 'Water Street'
    return name
```

```python
gb["Add_StNew"] = gb["Add_StName"].apply(lambda row: Add_StRe(str(row)))
gb.head()
```

```python
gb.Add_StNew.value_counts()
```

```python
gb.City.value_counts()
```

```python
def City_Re(x): 
    name = x
    if ('New York' in x):
        name = 'New York'
    return name
```

```python
gb["CityNew"] = gb["City"].apply(lambda row: City_Re(str(row)))
gb.head()
```
